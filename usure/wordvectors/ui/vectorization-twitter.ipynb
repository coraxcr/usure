# 1. Corpus:  Twitter 

```python
w2v_sw = w2v_dao.get_model("twitter.sw")
w2v_bg = w2v_dao.get_model("twitter.bw")
reporter_sw = StatisticsReporter(w2v_sw.wv)
reporter_bw = StatisticsReporter(w2v_bg.wv)
#twitter = get_model("tweets")
#cardellino = get_model("xxx")
```

## 1.1 Histograma

```python
reporter_sw.plot_histogram(numb_top_words=65)
```

## 1.2 Nube de palabras

```python
reporter_sw.plot_wordcloud(max_words=250)
```

## 1.3 Representacion vectorial en 2D

```python
numb_words = 500
reporter_sw.plot_scatter(numb_words=numb_words, num_dim=2)
```

```python
reporter_bw.plot_scatter(numb_words=numb_words, num_dim=2)
```
