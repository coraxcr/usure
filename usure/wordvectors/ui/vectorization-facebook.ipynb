# 1. Corpus:  Facebook2013 

```python
kvs_sw = kvs_rep.get("CorpusFBCR2013.txt.usu.sw.kvs")
kvs_bg = kvs_rep.get("CorpusFBCR2013.txt.usu.bw.kvs")
reporter_sw = StatisticsReporter(kvs_sw)
reporter_bw = StatisticsReporter(kvs_bg)
```

## 1.1 Histograma

```python
reporter_sw.plot_histogram(numb_top_words=65)
```

## 1.2 Nube de palabras

```python
reporter_sw.plot_wordcloud(max_words=250)
```

## 1.3 Representacion vectorial en 2D

```python
numb_words = 500
reporter_sw.plot_scatter(numb_words=numb_words, num_dim=2)
```

```python
reporter_bw.plot_scatter(numb_words=numb_words, num_dim=2)
```
